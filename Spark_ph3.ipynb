```python
import pandas as pd
import os

#Import various files and save as different datasets

xls_1 = pd.ExcelFile("./dataset-phase2/Target4.1 (1).xlsx")
xls_2 = pd.ExcelFile("./dataset-phase2/Target4.2.xlsx")
xls_3 = pd.ExcelFile("./dataset-phase2/Target4.3.xlsx")
xls_4 = pd.ExcelFile("./dataset-phase2/Target4.4 .xlsx")
xls_4c = pd.ExcelFile("./dataset-phase2/Target4.c.xlsx")
xls_5 = pd.ExcelFile("./dataset-phase2/Target4.5.xlsx")
xls_6 = pd.ExcelFile("./dataset-phase2/Target4.6.xlsx")

df_1 = pd.read_excel(xls_1, 'Data')
df_2 = pd.read_excel(xls_2, 'Data')
df_3 = pd.read_excel(xls_3, 'Data')
df_4 = pd.read_excel(xls_4, 'Data')
df_4c = pd.read_excel(xls_4c, 'Data')
df_5 = pd.read_excel(xls_5, 'Data')
df_6 = pd.read_excel(xls_6, 'Data')

df_meta = pd.read_excel(xls_1, 'Country - Metadata')
df_meta=df_meta.iloc[:,[0,2]]
df_meta.rename(columns = {'Code':'Country Code'}, inplace = True)
```

```python
#Concatenating all the dataframes

df = pd.concat([df_1,df_2,df_3,df_4,df_4c,df_5,df_6])
df.drop(df.columns[[4, 5, 21]], axis = 1 , inplace = True)
df
```

```python
#Deleting rows with less than three values

df = df[df.iloc[:,4:19].isnull().sum(axis=1) < 12]
df.shape
```

### Imputing with backward-fill 
Chosen imputator is bfill based on the step 1 & 2

```python
#Back filling NaN values
df=df.bfill(axis='rows')

#Filling NaN values with zero
df=df.fillna(0)

```

```python
# Simplifying year column names
cols = df.columns.tolist()
temp = cols[:4]
years = [i.split()[0] for i in cols[4:]]
cols = temp + years
cols
```

```python
# Changing df years columns names
df.columns = cols 
```

```python
#Adding Income Group Column to the datasets

df = pd.merge(df, df_meta, how="left" ,   on='Country Code', suffixes=( None,None))
df.head()
```

```python
df
```

```python
df = (df.drop(columns=['Series Code', 'Country Name', 'Country Code'])
        .groupby(['Series Name' , 'Income Group'] , as_index=False )
        .agg("mean")
)
```

```python
df_final = pd.DataFrame(columns = ["year"] + ["Income Group"] + df["Series Name"].unique().tolist())
df_final["year"] = [year for years in [[year]*4 for year in years] for year in years]
df_final["Income Group"] = [j for i in range(len(years)) for j in df["Income Group"].unique()]
df_final
```

```python
for year in years:
    for income in df["Income Group"].unique():
        df_final.loc[ (df_final["year"] == year) & (df_final["Income Group"] == income) , df_final.columns[2:]] = (
            df.loc[df["Income Group"] == income , year].tolist()
        )
```

```python
import pyspark
from pyspark.context import SparkContext
from pyspark.sql.session import SparkSession
sc = SparkContext('local')
spark = SparkSession(sc)
```

```python
df_final=spark.createDataFrame(df_final) 
```

```python
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler

df_features=df_final.drop("Income Group", "year")

vec_assembler = VectorAssembler(inputCols = df_features.columns, outputCol='features')
feature_data = vec_assembler.transform(df_final)
```

```python
#scaling the data
from pyspark.ml.feature import StandardScaler
scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures", withStd=True, withMean=False)
# Compute summary statistics by fitting the StandardScaler
scalerModel = scaler.fit(feature_data)
# Normalize each feature to have unit standard deviation.
final_data = scalerModel.transform(feature_data)
final_data['scaledFeatures','Income Group'].show()
```

```python
from pyspark.ml.feature import StringIndexer
indexer = StringIndexer(inputCol="Income Group", outputCol="labels")
final_data=indexer.fit(final_data).transform(final_data)
```

```python
final_data['scaledFeatures','labels'].show()
```

```python
#split data into training and testing
(trainingData, testData) = final_data.randomSplit([0.75, 0.25],seed = 11)
```

## Using DecisionTree as estimator for CrossVal

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder

# A CrossValidator requires an Estimator, a set of Estimator ParamMaps, and an Evaluator.

evaluator = MulticlassClassificationEvaluator(labelCol="labels", predictionCol="prediction", metricName="accuracy")
dt = DecisionTreeClassifier(labelCol="labels", featuresCol="scaledFeatures")
evaluator = MulticlassClassificationEvaluator(labelCol="labels", predictionCol="prediction", metricName="accuracy")
#ParamGridBuilder to construct a grid of parameters to search over
paramGrid = ParamGridBuilder() \
 .addGrid(dt.maxDepth, [3, 4, 5]) \
 .addGrid(dt.maxBins, [30, 40, 50]) \
 .build()
#declaring  crossvalidation object with k=10
crossval = CrossValidator(estimator=dt,
                          estimatorParamMaps=paramGrid,
                          evaluator=evaluator,
                          numFolds=10)
# Run cross-validation, and choose the best set of parameters.
cvModel = crossval.fit(trainingData)
#predict the test values

# Make predictions on test documents. cvModel uses the best model found (lrModel)
prediction = cvModel.transform(testData)
prediction.select("prediction", "labels", "scaledFeatures").show()
```

```python
#accuracy on test dataset
accuracy = evaluator.evaluate(prediction)
print("Test Error = %g" % (1.0 - accuracy))
```

```python
#best model
cvModel.bestModel
```

## Using RandomForest as estimator for CrossVal

```python
from pyspark.ml.classification import RandomForestClassifier

# A CrossValidator requires an Estimator, a set of Estimator ParamMaps, and an Evaluator.

rf = RandomForestClassifier(labelCol="labels", featuresCol="scaledFeatures")
evaluator = MulticlassClassificationEvaluator(labelCol="labels", predictionCol="prediction", metricName="accuracy")
#ParamGridBuilder to construct a grid of parameters to search over
paramGrid = ParamGridBuilder() \
 .addGrid(rf.maxDepth, [2, 4, 6]) \
 .addGrid(rf.maxBins, [15, 30, 45]) \
 .addGrid(rf.numTrees, [3, 4, 5]) \
 .build()
#declaring  crossvalidation object with k=10
crossval = CrossValidator(estimator=rf,
                          estimatorParamMaps=paramGrid,
                          evaluator=evaluator,
                          numFolds=10)
# Run cross-validation, and choose the best set of parameters.
cvModel = crossval.fit(trainingData)
#predict the test values

# Make predictions on test documents. cvModel uses the best model found (lrModel)
prediction = cvModel.transform(testData)
prediction.select("prediction", "labels", "scaledFeatures").show()
```

```python
#accuracy on test dataset
accuracy = evaluator.evaluate(prediction)
print("Test Error = %g" % (1.0 - accuracy))
```

```python
#best model
cvModel.bestModel
```

## Using MultilayerPerceptron as estimator for CrossVal

```python
from pyspark.ml.classification import MultilayerPerceptronClassifier

mpc = MultilayerPerceptronClassifier(labelCol="labels", featuresCol="scaledFeatures",maxIter=1000)

#ParamGridBuilder to construct a grid of parameters to search over
paramGrid = ParamGridBuilder() \
 .addGrid(mpc.layers, [[20, 3, 4, 3, 4], [20, 7, 5, 7, 4], [20, 12, 8, 12, 4]]) \
 .build()

#declaring  crossvalidation object with k=10
crossval = CrossValidator(estimator=mpc,
                          estimatorParamMaps=paramGrid,
                          evaluator=evaluator,
                          numFolds=10)

# Run cross-validation, and choose the best set of parameters.
cvModel = crossval.fit(trainingData)


# Make predictions on test documents. cvModel uses the best model found 
prediction = cvModel.transform(testData)
prediction.select("prediction", "labels", "scaledFeatures").show()
```

```python
#accuracy on test dataset
accuracy = evaluator.evaluate(prediction)
print("Test Error = %g" % (1.0 - accuracy))
```
